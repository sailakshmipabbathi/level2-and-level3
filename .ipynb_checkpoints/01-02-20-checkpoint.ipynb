# classification
## Logistic Regression Model
- It is a classification algorithm,used to predict binary outcomes.(0/1,yes/no,true/false,success/failure,head/tail e.t.c)

## Working of Logistic regression
- It measures the relationship between input(Independent variable) and Target(Dependent variable) by
estimating probabilities using sigmoid().


## Sigmoid function
- This function is an 'S' shaped curve that map the real-valued number(between 0 to 1).
- It can converts any line into curve.

## Formula
   p=1/1+e^-y   
   y=mx+c  => p=1/e^-y  
   
   p=1/1+e^-(mx+c)  => ln(p/1-p)=mx+c

```python
df=pd.read_csv("bid.csv")
df
```

```python
df.describe()
```

```python
X=df["Number"].values
Y=df["Win_Loss"].values
```

```python
plt.scatter(X,Y)
```

```python
n=len(X)
print(np.ndim(X))
X=X.reshape(n,1)
print(np.ndim(X))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression(C=1.0,solver="lbfgs",multi_class="ovr")   #ovr=one versus rest
log_reg.fit(X,Y)
log_predict=log_reg.predict(X)
log_predict
```

```python
print(log_reg.predict_proba([[200]]))
```

```python
plt.plot(X,log_predict)
```

```python
plt.scatter(X,Y)
plt.plot(X,log_predict)
```
