# Matplotlib

"Matplotlib is a python 2d plotting"

## types of plotting in matplot
- lines
- Bar charts
- scatter plot
- pie chart 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

# Line

```python
plt.plot([1,2,3,4,5],[2,6,10,12,16])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plotting")
plt.show()
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label="line",color='b')
plt.plot(x,x**2,label="square",color='g')
plt.title('line and square \nplotting')
plt.legend()
plt.show()

```

# bar graph

```python
plt.bar([1,2,3,4],[2,3,4,5])
```

```python
plt.plot([1,2,3,4],[2,3,4,5],'y*')
```

```python
car1=[0,1,2,3,4]
car1_kn=[34,56,78,90,67]
car2=[2,3,4,5,6]
car2_kn=[45,66,77,88,99]
plt.bar(car1,car1_kn,label="benz",color='b')
plt.bar(car2,car2_kn,label="toyata",color='r')
plt.xlabel("no of days")
plt.ylabel("kilometers")
plt.legend()
plt.show()
```

# areaplot

```python
days=[1,2,3,4,5]
working=[8,8,5,9,10]
sleeping=[8,10,7,11,12]
plt.stackplot(days,working,sleeping,colors=['r','y'],labels=["working","sleeping"])
plt.legend()
plt.title("area")
plt.show()
```

# piechart

```python
days=[1,2,3,4,5]
working=[8,8,5,9,10]
sleeping=[8,10,7,11,12]
slices=[4,6,8]
variables=["days","working","sleeping"]
plt.pie(slices,labels=variables,colors=['r','b','g'],startangle=90,autopct='%1.2f%%')
plt.show()
```
